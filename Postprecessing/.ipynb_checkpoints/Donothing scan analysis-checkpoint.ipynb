```python
from collections import defaultdict

# Initialize a dictionary to store the results
results = {}

# Initialize a defaultdict to store unique domains per category
unique_domains_category = defaultdict(lambda: defaultdict(set))

# Iterate through each category
for category in categories_to_count:
    # Filter rows where the category is present and disconnect_category is not None
    df_filtered = flowReport[flowReport['disconnect_category'].apply(lambda x: x is not None and category in x)]
    
    # Iterate over each row in the filtered DataFrame
    for index, row in df_filtered.iterrows():
        entropy_level = row['entropy_level']
        disconnect_category = category
        
        # Retrieve the set of unique domains for the specific category and entropy level
        domains_set = row['matching_domains']
        
        # Update the set of unique domains for the specific category and entropy level
        unique_domains_category[disconnect_category][entropy_level].update(domains_set)

# Process the unique domains for each category
for disconnect_category, entropy_levels in unique_domains_category.items():
    unique_domain_counts = defaultdict(int)
    for entropy_level, domains_set in entropy_levels.items():
        unique_domain_counts[entropy_level] = len(domains_set)
    
    # Store the counts in the results dictionary
    results[disconnect_category] = pd.DataFrame(unique_domain_counts.items(), columns=['entropy_level', 'unique_domains_count'])

    # Print the count of detected domains for each entropy level within the category
    print(f"Disconnect Category: {disconnect_category}")
    for entropy_level, count in entropy_levels.items():
        print(f"Entropy Level: {entropy_level}")
        print(f"Count of Unique Domains: {count}")
        print("\n")

```
